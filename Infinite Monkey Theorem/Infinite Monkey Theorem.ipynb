```python
import random
import string
```

```python
sen = 'methinks it is like a weasel'
l = list(string.ascii_lowercase)
```

```python
def fun(s):
    i = 0
    s1 = ''
    while(i<8):
        temp = list(random.choice(l))
        s1 = s1+ temp[0]
        i = i+1

    i = 0
    s2 = ''
    while(i<2):
        temp = list(random.choice(l))
        s2 = s2+ temp[0]
        i = i+1

    i = 0
    s3 = ''
    while(i<2):
        temp = list(random.choice(l))
        s3 = s3+ temp[0]
        i = i+1

    i = 0
    s4 = ''
    while(i<4):
        temp = list(random.choice(l))
        s4 = s4+ temp[0]
        i = i+1

    i = 0
    s5 = ''
    while(i<1):
        temp = list(random.choice(l))
        s5 = s5+ temp[0]
        i = i+1

    i = 0
    s6 = ''
    while(i<6):
        temp = list(random.choice(l))
        s6 = s6+ temp[0]
        i = i+1
    
    return (s1+' '+s2+' '+s3+' '+s4+' '+s5+' '+s6)
        
    

```

```python
def check(original_sen,output_sen):
    f = True #flag to get out of while loop
    case_number = 1 #to know how many cases have been generated by random
    while(f):
        matched = 0
        output_sen = fun(sen).split()

        t = len(sen.split()[0]) + len(sen.split()[1]) +len(sen.split()[2]) +len(sen.split()[3])+len(sen.split()[4]) +len(sen.split()[5])
        templ = sen.split()

        for i in templ:
            for j in i:
                if j == output_sen[templ.index(i)][i.index(j)]:
                    matched = matched+1
        per_matched = (matched/t)*100
        if per_matched>20:
            print('20 percent match found')
            print(fun(sen))
            print(sen)
            print(case_number)
            print('\n')
        elif per_matched>30:
            print('30 percent match found')
            print(fun(sen))
            print(sen)
            print(case_number)
            print('\n')
        elif per_matched==100:
            print('We got ittttt!')
            f = False
        case_number = case_number+1
             
    
    
    
```
