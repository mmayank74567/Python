```python
from random import choice
def get_month():
    months = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'] #12 months range
    v = choice(months)
    return v
```

```python
match = 0
x = 0
for i in range(10):
  x = x+1
  A = get_month()
  B = get_month()
  C = get_month()
  D = get_month()
  E = get_month()

  if (A==B or A==C or A==D  or A==E  or B==C  or B==D  or B==E  or C==D  or C==E  or D==E):
    match = match+1
    print('Common birthday found at',x, 'iteration ',[A,B,C,D,E])
print('No of times a match was found',match)
```
